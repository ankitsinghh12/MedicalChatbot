```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
fdist = FreqDist()
for word in hamlet:
    fdist[word.lower()]+=1
fdist
```

```python
fdist['hamlet'] # the frequency of the word 'hamlet'
```

```python
len(fdist) # number of distinct words
```

```python
fdist_top10 = fdist.most_common(10)
fdist_top10
```

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
quote = "The greatest glory in living lies not in never falling, but in rising every time we fall."
quote_tokens = nltk.word_tokenize(quote)
quote_tokens
```

```python
quote_bigrams = list(nltk.bigrams(quote_tokens))
quote_trigrams = list(nltk.trigrams(quote_tokens))
quote_ngrams = list(nltk.ngrams(quote_tokens, 5))
quote_ngrams
```

```python
import re
punctuation = re.compile(r'[-.?!,:;()\[\]\'|0-9]')
post_punctuation = []
for word in hamlet:
    w = punctuation.sub("", word)
    if len(w) > 0:
        post_punctuation.append(w)
        
post_punctuation
```

```python
message = "Tim is a natural when it comes to drawing."
message_tokens = word_tokenize(message)
for token in message_tokens:
    print(nltk.pos_tag([token]))
```

```python
from nltk import ne_chunk
ne_message = "The US President stays in the White House."
ne_tokens = word_tokenize(ne_message)
ne_tags = nltk.pos_tag(ne_tokens)
ne_ner = ne_chunk(ne_tags)
```

```python
print(ne_ner)
```

```python
new = "The big cat ate the little mouse who was after fresh cheese"
new_tokens = nltk.pos_tag(word_tokenize(new))
new_tokens
```

```python
chunk_parser = nltk.RegexpParser(r"NP: {<DT>?<JJ>*<NN>}")
chunk_result = chunk_parser.parse(new_tokens)
chunk_result
```
